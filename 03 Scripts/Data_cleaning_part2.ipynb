## 03. Wrangling the data to have a common column with the geoJSON file

```python
# renaming the 'locality/district' column to 'locality_district'

df.rename(columns={'locality/district':'locality_district'},inplace=True)
```

```python
# creating a new column named 'district' by associating each locality with its respective district for the geoJSON mapping

district = []

for x in range(len(df.locality_district)):
    if df.locality_district[x] in ['Charlottenburg_Charlottenburg','Wilmersdorf_Wilmersdorf','Schmargendorf_Wilmersdorf','Grunewald_Wilmersdorf']:
        district.append('Charlottenburg-Wilmersdorf')
    elif df.locality_district[x] in ['Friedrichshain_Friedrichshain','Kreuzberg_Kreuzberg']:
        district.append('Friedrichshain-Kreuzberg')
    elif df.locality_district[x] in ['Friedrichsfelde_Lichtenberg','Lichtenberg_Lichtenberg','Alt_Hohenschönhausen_Hohenschönhausen','Neu_Hohenschönhausen_Hohenschönhausen','Karlshorst_Lichtenberg','Falkenberg_Hohenschönhausen','Malchow_Hohenschönhausen','Rummelsburg_Lichtenberg']:
        district.append('Lichtenberg')
    elif df.locality_district[x] in ['Hellersdorf_Hellersdorf','Marzahn_Marzahn','Biesdorf_Marzahn','Mahlsdorf_Hellersdorf','Kaulsdorf_Hellersdorf']:
        district.append('Marzahn-Hellersdorf')
    elif df.locality_district[x] in ['Mitte_Mitte','Tiergarten_Tiergarten','Wedding_Wedding']:
        district.append('Mitte')
    elif df.locality_district[x] in ['Neukölln_Neukölln','Buckow_Neukölln','Britz_Neukölln','Rudow_Neukölln']:
        district.append('Neukölln')
    elif df.locality_district[x] in ['Prenzlauer_Berg_Prenzlauer_Berg','Weißensee_Weißensee','Pankow_Pankow','Niederschönhausen_Pankow','Karow_Weißensee','Französisch_Buchholz_Pankow','Rosenthal_Pankow','Heinersdorf_Weißensee','Buch_Pankow','Blankenburg_Weißensee']:
        district.append('Pankow')
    elif df.locality_district[x] in ['Reinickendorf_Reinickendorf','Tegel_Reinickendorf','Waidmannslust_Reinickendorf','Wittenau_Reinickendorf','Hermsdorf_Reinickendorf','Frohnau_Reinickendorf','Heiligensee_Reinickendorf','Konradshöhe_Reinickendorf','Lübars_Reinickendorf']:
        district.append('Reinickendorf')
    elif df.locality_district[x] in ['Spandau_Spandau','Staaken_Spandau','Siemensstadt_Spandau','Kladow_Spandau','Haselhorst_Spandau','Gatow_Spandau']:
        district.append('Spandau')
    elif df.locality_district[x] in ['Lichterfelde_Steglitz','Steglitz_Steglitz','Zehlendorf_Zehlendorf','Lankwitz_Steglitz','Dahlem_Zehlendorf','Wannsee_Zehlendorf','Nikolassee_Zehlendorf']:
        district.append('Steglitz-Zehlendorf')
    elif df.locality_district[x] in ['Schöneberg_Schöneberg','Tempelhof_Tempelhof','Lichtenrade_Tempelhof','Marienfelde_Tempelhof','Mariendorf_Tempelhof','Friedenau_Schöneberg']:
        district.append('Tempelhof-Schöneberg')
    else:
        district.append('Treptow-Köpenick')

df['district']=district
```
