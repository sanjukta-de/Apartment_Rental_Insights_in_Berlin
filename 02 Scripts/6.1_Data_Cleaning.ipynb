## 04. Cleaning the data

```python
df_berlin.info()
```

```python
# checking the missing values

df_berlin.isnull().sum()
```

```python
# checking the percentage of the missing values

null_percent=df_berlin.isnull().sum()/len(df_berlin)*100
```

```python
null_percent
```

```python
# dropping all the columns having null values more than 30%

df_berlin = df_berlin.drop(columns = df.columns[null_percent > 30])
```

```python
df_berlin.shape
```

```python
pd.options.display.max_columns=None
```

```python
df_berlin.head()
```

```python
# since I want to analyze the rental price(totalRent), so would delete all the rows having null 'totalRent' value

df_berlin.dropna(subset=['totalRent'],inplace=True)
```

```python
df_berlin.shape
```

```python
# dropping all the columns not required for my analysis

df_berlin.drop(columns=['picturecount','pricetrend','scoutId','geo_bln','yearConstructedRange','houseNumber','geo_krs','street','baseRentRange','noRoomsRange','livingSpaceRange','description','facilities','date'],inplace=True)
```

```python
df_berlin.shape
```

```python
df_berlin.head(10)
```

```python
# checking the Mean value of all the numerical columns

df_berlin._get_numeric_data().mean()
```

```python
# replacing the NaN values of the numerical columns with the Mean values

df_berlin.fillna(df_berlin._get_numeric_data().mean(), inplace=True)
```

```python
# replacing the NaN values of all the categorical columns with the Mode 

for col in df_berlin.columns:
    if df_berlin[col].dtype == 'object':
        df_berlin[col] = df_berlin[col].fillna(df_berlin[col].mode()[0])
```

```python
# checking the null values again

df_berlin.isnull().sum()
```

```python
df_berlin.describe()
```

```python
# checking for any duplicates

df_berlin.duplicated().sum()
```

```python
# deleting the duplicates

df_berlin.drop_duplicates(inplace=True)
```

```python
df_berlin.shape
```

```python
# deleting the Outliers

for cols in df_berlin.columns:
    if df_berlin[cols].dtype == 'int64' or df_berlin[cols].dtype == 'float64':
        upper_range = df_berlin[cols].mean() + 3 * df_berlin[cols].std()
        lower_range = df_berlin[cols].mean() - 3 * df_berlin[cols].std()
        
        indexes = df_berlin[(df_berlin[cols] > upper_range)|(df_berlin[cols] < lower_range)].index
        df_berlin = df_berlin.drop(indexes) 
```

```python
df_berlin.shape
```

```python
df_berlin.describe()
```

```python
pd.options.display.max_rows=None
```

```python
# checking the'totalRent' column to see the outliers

df_berlin['totalRent'].value_counts().unique
```

```python
# there seems to be 3 totalRent less than 200€ which seems to be unlikely, so deleting them

rent_clean=df_berlin[(df_berlin['totalRent']>200) & (df_berlin['totalRent']<4000)]
```

```python
rent_clean.describe()
```

```python
df_berlin=rent_clean
```

```python
df_berlin.shape
```

```python
# checking the baseRent column to see the outliers

df_berlin['baseRent'].value_counts().unique
```

```python
# there seems to be 1 baseRent less than 50€ which seems to be unlikely, so deleting it

df_berlin=df_berlin[(df_berlin['baseRent']>50) & (df_berlin['baseRent']<4000)]
```

```python
df_berlin.describe()
```

```python
df_berlin.shape
```
